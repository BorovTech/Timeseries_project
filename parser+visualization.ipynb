```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f = open("markdown_to_text.txt", "r")

to_map = []

for line in f:
    temp = list(map(str, line.split("|")))
    to_map.append(list([temp[1], temp[3], temp[4]]))

silhouette_map = dict()
db_score_map = dict()


for x in to_map:
    silhouette_map[x[0]] = float(x[1])
    db_score_map[x[0]] = float(x[2])

f.close()
```

```python
# silhouette_plot = dict(sorted(silhouette_map.items(), key=lambda item: item[1]))
silhouette_plot_list = list(silhouette_map.values())
```

```python
plot = sns.scatterplot(y=silhouette_plot_list, x=list(range(len(silhouette_plot_list))))
plot.set_title("Silhouette score unsorted")
```

```python
silhouette_plot1 = dict(sorted(silhouette_map.items(), key=lambda item: item[1]))
silhouette_plot_list1 = list(silhouette_plot1.values())
plot1 = sns.scatterplot(y=silhouette_plot_list1, x=list(range(len(silhouette_plot_list1))))
plot1.set_title("Silhouette score sorted")
```

```python
db_plot_list = list(db_score_map.values())
plot_db = sns.scatterplot(y=db_plot_list, x=list(range(len(db_plot_list))))
plot_db.set_title("DB score unsorted")
```

```python
db_plot1 = dict(sorted(db_score_map.items(), key=lambda item: item[1]))
db_plot_list1 = list(db_plot1.values())
plot_db1 = sns.scatterplot(y=db_plot_list1, x=list(range(len(db_plot_list1))))
plot_db1.set_title("DB score sorted")
```
