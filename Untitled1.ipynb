```python
def small_patterns(arr):
    #if arr[i + 1] > arr[i] -> 1
    #else 0
    output = []
    for i in range(len(arr)-1):
        if arr[i + 1] >= arr[i]:
            output.append(1)
        else:
            output.append(0)
    return output

def check_directions(motif, series_window):
    if small_patterns(motif) == small_patterns(series_window):
        return True
    else:
        return False
    
def check_boundaries_euclidean(motif, series_window, EPS=0.05):
    dist = euclidean(motif, series_window)
    if dist < EPS:
        return True
    return False
```

```python
def clustering_by_tema(clustering_data):
    metas = []
    ans = [0]*len(clustering_data)
    for series in clustering_data:
        metas.append(small_patterns(series))
    #le_dict = dict((tuple(el),[]) for el in metas)
    visited = []
    for i in range(len(clustering_data)):
        if metas[i] not in visited:
            visited.append(metas[i])
            for j in range(len(clustering_data)):
                    if metas[j] == metas[i]:
                        #le_dict[metas[i]].append(clustering_data(j))
                        ans[j] = i
    return ans
```
