```python
def small_patterns(arr):
    #if arr[i + 1] > arr[i] -> 1
    #else 0
    output = []
    for i in range(len(arr)-1):
        if arr[i + 1] >= arr[i]:
            output.append(1)
        else:
            output.append(0)
    return output

def check_directions(motif, series_window):
    if small_patterns(motif) == small_patterns(series_window):
        return True
    else:
        return False
    
def check_boundaries_euclidean(motif, series_window, EPS=0.05):
    dist = euclidean(motif, series_window)
    if dist < EPS:
        return True
    return False
```

```python
def clustering_by_tema(clustering_data):
    metas = []
    ans = [0]*len(clustering_data)
    for series in clustering_data:
        metas.append(small_patterns(series))
    #le_dict = dict((tuple(el),[]) for el in metas)
    visited = []
    for i in range(len(clustering_data)):
        if metas[i] not in visited:
            visited.append(metas[i])
            for j in range(len(clustering_data)):
                    if metas[j] == metas[i]:
                        #le_dict[metas[i]].append(clustering_data(j))
                        ans[j] = i
    return ans
```

```python
def Ivanov_clustering(clustering_data):  # data - matrix with series
    direction_clusternig = clustering_by_tema(clustering_data)
    clusters = [0]*len(clustering_data)
    for direction_cluster_number in set(direction_clusternig):
        cluster_indexes = [i for i,x in enumerate(direction_clusternig) if x==direction_cluster_number]
        direction_cluster = [clustering_data[x] for x in cluster_indexes]
        new_wishart = Wishart(2, 1)
        clustering = new_wishart.fit(np.array(direction_cluster), workers=1, batch_weight_in_gb=50)
        clustering_complete = []
        for x in range(len(clustering)):
            if clustering[x] != 0:
                clustering_complete.append(direction_cluster_number*1000 + clustering[x])
            else:
                clustering_complete.append(0)
        for index in range(len(cluster_indexes)):
            clusters[cluster_indexes[index]] = clustering_complete[index]
    return clusters
```
